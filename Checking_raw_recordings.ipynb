```python
import numpy as np
import os
import mne
import matplotlib as plt
plt.use('TkAgg')
```

# Get All File Path

```python
def get_file_paths(directory):
    file_paths = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            file_paths.append(file_path)
    return file_paths

folder_path = input("path to wakefulness files: ")
all_file_paths = get_file_paths(folder_path)
```

# Plot raw data region by region

```python
for rec in all_file_paths:
    raw_data = mne.io.read_raw_edf(rec)
    print(raw_data.info)
    raw_data.plot()
```
