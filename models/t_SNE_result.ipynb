```python
import numpy as np
import os
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python

# Directory containing .npz files
file_path = "/Users/wuruoyu/Sleep-Staging/data/train_data2_noCoordinates.npz"
data = np.load(file_path)
X = data['X'][:,0:3000]
y = data['y']
```

```python

# Optional: Normalize the data (e.g., MinMax scaling or standardization)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Apply t-SNE to reduce dimensionality to 2D
tsne = TSNE(n_components=2, perplexity=4, n_iter=400, random_state=42)
tsne_results = tsne.fit_transform(X)  # Shape: (1300+, 2)

# Visualize the t-SNE results
plt.figure(figsize=(10, 8))
scatter = plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=y, cmap='viridis', s=10)
plt.colorbar(scatter, label="Labels")
plt.title("t-SNE Visualization of Time-Series Data")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
