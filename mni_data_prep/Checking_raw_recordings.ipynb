```python
import numpy as np
import os
import mne
import matplotlib as plt
import pandas as pd
import csv
plt.use('TkAgg')
```

# Get All File Path

```python
all_states = {1:'W',2:'R',3:'N2',4:'N3'}

def get_file_paths(directory):
    file_paths = []
    file_names = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            file_paths.append(file_path)
            file_names.append(file)
    return file_paths,file_names

# folder_path = input(f"path to {cur_state} files: ")
# all_file_paths,all_file_names = get_file_paths(folder_path)
```

# Plot&Extract channel info raw data region by region

```python
chan_info = pd.DataFrame(columns=['ch_name','region','W','R','N2','N3'])
chan_data = np.empty((0,13603))

def fill_chan_info(chan_info_df, all_file_names, cur_state):
    for i in range(len(all_file_names)):
        fp = all_file_paths[i]
        fn = all_file_names[i]
        raw = mne.io.read_raw_edf(fp)
        # raw_data = raw.get_data()
        for j in range(len(raw.info['ch_names'])):
            ch = raw.info['ch_names'][j]
            print(ch)
            try:
                row_no = chan_info[chan_info['ch_name'] == ch].index[0]
                chan_info.loc[row_no,cur_state] = 1
            except:
                chan_info.loc[len(chan_info),'ch_name'] = ch
                chan_info.loc[len(chan_info)-1,'region'] = fn[:-6]
                chan_info.loc[len(chan_info)-1,f'{cur_state}'] = 1
            # cur_data = np.concatenate((np.array([ch,cur_state,fn[:-5]]),raw_data[j]))
            # chan_data = np.vstack((chan_data,cur_data))
        # print(raw_data.info)
        # raw_data.plot()
    return chan_info

for state_no in range(1,5):
    cur_state = all_states[state_no]
    folder_path = input(f"path to {cur_state} files: ")
    all_file_paths,all_file_names = get_file_paths(folder_path)
    chan_info = fill_chan_info(chan_info,all_file_names,cur_state)

```
